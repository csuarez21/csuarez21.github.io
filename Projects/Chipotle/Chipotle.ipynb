<h1>Set up & data exploration
   

```python
df = pd.read_excel("Chipotle.xlsx")
df = df.iloc[: , :-1] 
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
### There are misspelled words in top1
df["top1"].value_counts()
```

```python
# Changing all strings to lowercase then removing any non alphabetic characters and spaces
df['top1'] = df['top1'].str.lower().replace('[^a-zA-Z0-9]', '', regex=True).str.strip().tolist()

# Correcting misspelled words
df.loc[df['top1'].str.contains('fila',na=False), 'top1'] = 'chickfila'
df.loc[df['top1'].str.contains('donalds|mcdondalds',na=False), 'top1'] = 'mcdonalds'
df.loc[df['top1'].str.contains('panera',na=False), 'top1'] = 'panera'
df.loc[df['top1'].str.contains('chip',na=False), 'top1'] = 'chipotle'
df.loc[df['top1'].str.contains('innoutkebobshop',na=False), 'top1'] = 'innout' # Change the row where a participant answered 2 restaurants instead of 1

df["top1"].value_counts()
```

```python
### Correct Chipolete survey answers exceeds range
for column in ['importantconvenience', 'importantvariety', 'importantprice', 'importanthealthy', 'importanttaste', 'importantambience', 'chipotleconvenient', 'chipotlevariety', 'chipotleprice', 'chipotlehealthy', 'chipotletaste', 'chipotleambience']:
    df[column] = df[column].replace([6.0],5.0)
```

```python
df[['importantconvenience', 'importantvariety', 'importantprice',
       'importanthealthy', 'importanttaste', 'importantambience',
       'chipotleconvenient', 'chipotlevariety', 'chipotleprice',
       'chipotlehealthy', 'chipotletaste', 'chipotleambience']].describe()
```

```python
#### Check null data
df.isnull().sum()
```

```python
#### remove rows that are missing more than 75% survey entries  (drop more than 75% NAN)
perc = 75.0 
min_count =  int(((100-perc)/100)*df.shape[1] + 1)
df = df.dropna( axis=0, thresh=min_count)
```

```python
### Fill Na with median data

df = df.fillna(df.median())
df
```

<h1>Clustering KMeans
    
    

```python
### First we have to convert the categorical data / string to numbers for our model to process
### We can use one hot encoder or pd get dummies
### since this is unsupervised learning we don't have to worry about data being missed so pd.getdummies will work 

df = pd.get_dummies(df)
df
```

```python
##### Pick the correct number of cluster using elbow/knee method
### Track decrease in SSD (Sum of the squared distances) vs increase in clusters
from sklearn.cluster import KMeans


ssd = []

### Because we want each cluster to have at least 20 rows the maximum cluster we will loop is 353 / 20 = 17
for i in range (2,17):
    model = KMeans(n_clusters= i)
    model.fit(df)
    
    ssd.append(model.inertia_)   ## SSD point --> cluster center
    
```

```python
plt.figure(figsize=(12,6), dpi=200)
plt.xlabel('# of Clusters')
plt.ylabel('SSD (Sum of the squared distances)')
plt.plot(range(2,17),ssd,'o--');
```

```python
pd.Series(ssd).diff
```

```python
# we choose 6 as our number of clusters
```

```python
model = KMeans(n_clusters= 6, random_state=42)
cluster_labels = model.fit_predict(df)
```

```python
cluster_labels
```

```python
df['Cluster'] = cluster_labels
```

<h1>Clustering Hierachical 

```python
### There are many similarity Metrics options: Euclidean Distance, Manhattan, cosine.... For simple data set like ours we will use the Euclidean Distance method to measure the distance between two points

df2 = df.loc[:, df.columns != 'Cluster']
```

```python
#### use a minMax Scaler       sqrt( (x1_a - x1_b)^2 )
from sklearn.preprocessing import MinMaxScaler

mScaler = MinMaxScaler()

X = mScaler.fit_transform(df2)
X = pd.DataFrame(X, columns=df2.columns)
```

```python
plt.figure(figsize=(6,4), dpi=200)
sns.clustermap(X, cmap='viridis')
```

```python
from sklearn.cluster import AgglomerativeClustering

modelH = AgglomerativeClustering(n_clusters = None, distance_threshold = 0)
```

```python
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster import hierarchy

linkage_martix = hierarchy.linkage(modelH.children_)
```

```python
## if we join two clusters, what is the distance between and how many points are under it
linkage_martix
```

```python
np.sqrt(len(X.columns))    ### This is our therootical max distance between two points
```
